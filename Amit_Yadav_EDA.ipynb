```python
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
```

```python
# Assuming you have the customers, products, and transactions DataFrames already loaded
transactions_customers = transactions.merge(customers, on="CustomerID")
merged_data = transactions_customers.merge(products, on="ProductID")
```

```python
merged_data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Region and sum TotalValue
region_revenue = merged_data.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=region_revenue.index, y=region_revenue.values, palette="viridis")
plt.title("Revenue by Region", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Total Revenue", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by Category and sum TotalValue
category_revenue = merged_data.groupby('Category')['TotalValue'].sum().sort_values()

# Plot
plt.figure(figsize=(8, 6))
category_revenue.plot(kind='barh', color='skyblue')
plt.title("Revenue by Product Category", fontsize=14)
plt.xlabel("Total Revenue", fontsize=12)
plt.ylabel("Category", fontsize=12)
plt.show()
```

```python
# Group by CustomerID and sum TotalValue
top_customers = merged_data.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(8, 6))
top_customers.plot(kind='barh', color='teal')
plt.title("Top 10 Customers by Revenue", fontsize=14)
plt.xlabel("Total Revenue", fontsize=12)
plt.ylabel("CustomerID", fontsize=12)
plt.gca().invert_yaxis()  # Ensure the highest value is at the top
plt.show()
```

```python
# Extract signup month
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
customers['SignupMonth'] = customers['SignupDate'].dt.to_period('M')

# Group by SignupMonth
signup_trends = customers.groupby('SignupMonth').size()

# Plot
plt.figure(figsize=(10, 6))
signup_trends.plot(kind='line', marker='o', color='purple')
plt.title("Customer Signups Over Time", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Signups", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Extract transaction month
merged_data['TransactionDate'] = pd.to_datetime(merged_data['TransactionDate'])
merged_data['TransactionMonth'] = merged_data['TransactionDate'].dt.to_period('M')

# Group by TransactionMonth
transaction_trends = merged_data.groupby('TransactionMonth').size()

# Plot
plt.figure(figsize=(10, 6))
transaction_trends.plot(kind='line', marker='o', color='orange')
plt.title("Transactions Over Time", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# 5. Product Popularity
product_sales = merged_data.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 6))
product_sales.plot(kind='bar', color='purple')
plt.title("Top 10 Most Sold Products", fontsize=14)
plt.xlabel("Product Name", fontsize=12)
plt.ylabel("Quantity Sold", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot for numeric columns in transactions
numeric_cols = ['Quantity', 'TotalValue', 'Price']
for col in numeric_cols:
    sns.boxplot(data=transactions, x=col)
    plt.title(f'Box Plot for {col}')
    plt.show()


from scipy.stats import zscore

# Compute Z-scores for numeric columns
z_scores = transactions[numeric_cols].apply(zscore)

# Identify rows with any Z-score > 3 (or < -3)
outliers_zscore = transactions[(z_scores > 3).any(axis=1)]
print("Outliers based on Z-score:\n", outliers_zscore)


# IQR method for outlier detection
for col in numeric_cols:
    Q1 = transactions[col].quantile(0.25)
    Q3 = transactions[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers_iqr = transactions[(transactions[col] < lower_bound) | (transactions[col] > upper_bound)]
    print(f"Outliers in {col} based on IQR:\n", outliers_iqr)
```

```python
# Pearson correlation for numerical columns
correlation_matrix = transactions[numeric_cols].corr(method='pearson')
print("Pearson Correlation Matrix:\n", correlation_matrix)


# Spearman correlation for numerical columns
spearman_correlation = transactions[numeric_cols].corr(method='spearman')
print("Spearman Correlation Matrix:\n", spearman_correlation)


import numpy as np

# Heatmap for Pearson Correlation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap for Pearson Correlation')
plt.show()

# Heatmap for Spearman Correlation
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap for Spearman Correlation')
plt.show()
```
