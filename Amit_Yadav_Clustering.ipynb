```python
import pandas as pd

# Load the two datasets
transactions = pd.read_csv('Transactions.csv')
customers = pd.read_csv('Customers.csv')

# Merge the datasets on CustomerID
merged_data = pd.merge(transactions, customers, on='CustomerID', how='left')

# Check the merged dataset structure
print(merged_data.head())
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Assuming your merged_data DataFrame is already loaded
# Convert necessary columns to datetime
merged_data['TransactionDate'] = pd.to_datetime(merged_data['TransactionDate'])
merged_data['SignupDate'] = pd.to_datetime(merged_data['SignupDate'])

# Step 1: Calculate derived features
current_date = datetime(2024, 12, 31)  # Reference date for recency calculation
customer_features = merged_data.groupby('CustomerID').agg(
    total_spending=('TotalValue', 'sum'),
    transaction_count=('TransactionID', 'count'),
    last_transaction=('TransactionDate', 'max'),
    region=('Region', 'first')  # Assuming region remains constant per customer
).reset_index()

# Calculate recency (days since last transaction)
customer_features['recency'] = (current_date - customer_features['last_transaction']).dt.days

# One-hot encode Region
customer_features = pd.get_dummies(customer_features, columns=['region'], drop_first=True)

# Drop unnecessary columns
customer_features.drop(columns=['CustomerID', 'last_transaction'], inplace=True)

# Step 2: Normalize features
scaler = StandardScaler()
features = [col for col in customer_features.columns if col != 'cluster']
scaled_features = scaler.fit_transform(customer_features)

# Step 3: Hyperparameter tuning for K-Means
db_scores = []
cluster_range = range(2, 11)  # Test cluster counts from 2 to 10

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(scaled_features)
    db_index = davies_bouldin_score(scaled_features, labels)
    db_scores.append(db_index)
    print(f"Clusters: {n_clusters}, DB Index: {db_index}")

# Find the optimal number of clusters
optimal_clusters = cluster_range[db_scores.index(min(db_scores))]
print(f"Optimal number of clusters: {optimal_clusters} (DB Index: {min(db_scores)})")

# Plot the DB Index for each cluster count
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, db_scores, marker='o', linestyle='-', color='b')
plt.title('Hyperparameter Tuning: DB Index vs. Number of Clusters', fontsize=16)
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
plt.xticks(cluster_range)
plt.grid()
plt.show()

# Step 4: Fit the final model with the optimal number of clusters
kmeans_final = KMeans(n_clusters=optimal_clusters, random_state=42)
customer_features['cluster'] = kmeans_final.fit_predict(scaled_features)

# Step 5: Visualize clusters using PCA
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)
customer_features['pca_x'] = pca_features[:, 0]
customer_features['pca_y'] = pca_features[:, 1]

plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=customer_features,
    x='pca_x', y='pca_y',
    hue='cluster', palette='Set2', s=100, alpha=0.8
)
plt.title('Customer Clusters (PCA Reduced Dimensions)', fontsize=16)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.grid()
plt.show()

# Step 6: Display cluster summary
numeric_features = ['total_spending', 'transaction_count', 'recency'] + [col for col in customer_features.columns if col.startswith('region_')]
cluster_summary = customer_features.groupby('cluster')[numeric_features].mean()
print("Cluster Summary:")
print(cluster_summary)

# Save the results for further use
customer_features.to_csv('customer_clustering_results.csv', index=False)
```
